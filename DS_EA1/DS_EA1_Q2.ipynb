```python
import pandas as pd
import numpy as np
import math
# configure float formatting
pd.options.display.float_format = "{:.2f}".format
```

## Load the (countries.csv)[https://github.com/WHPAN0108/BHT-DataScience-S23/blob/main/python-DS/country.csv] directly via URL import into your panda data frame!


```python
countries_url = "https://github.com/WHPAN0108/BHT-DataScience-S23/raw/main/python-DS/country.csv"
countries = pd.read_csv(countries_url, delimiter=',')
```

```python
countries
#print(countries["Population"])
```

```python
def DisplayDescriptiveStatistics(dataset, columns, descrColumns):
    def resolveDescrColumn(descCol):
        def resolveColumn(column):
            return descCol[2](dataset[column])
        lst = [descCol[0]] + list(map(resolveColumn, columns)) + [descCol[1]]
        return tuple(lst)
    return list(map(resolveDescrColumn, descrColumns))

#def DisplayDescriptiveStatisticsOneLiner(dataset, columns, descrColumns):
#    return map(descrColumns, lambda descCol: [[descCol[0]] + map(columns, lambda column:descCol[2](dataset[column]))] + [descCol[1]])
```

## Display descriptive statistics for the numerical column (count, mean, std, min, 25%, 50%, 75%, max) HINT: describe

```python
numericColumns = ["Population", "Area", "GDP"]
descriptiveColumns = [
    ("count", "the number of data points", np.size),
    ( "mean", "the arithmetic mean of the values", np.mean),
    ("std", "the standard deviation", np.std), 
    ("min", "the minimal value", np.min),
    ("25%", "first quartile", lambda x:np.quantile(x, 0.25)),
    ("50%", "second quartile", lambda x:np.quantile(x, 0.50)),
    ("75%", "third quartile", lambda x:np.quantile(x, 0.75)),
    ("max", "the minimal value", np.max)    
    ]

descriptiveStatistics = DisplayDescriptiveStatistics(countries, numericColumns, descriptiveColumns)

pd.DataFrame(descriptiveStatistics, columns=[["STAT"] + numericColumns + ["DESCRIPTION"]])
```

## Show the last 4 rows of the data frame.


```python
pd.DataFrame(countries.values[-4:], columns=countries.columns)
```

## Show all the rows of countries that have the EURO

```python
countries.where(lambda data: data["Currency"] == "EUR").dropna()
```

## Show only name and Currency in a new data frame

```python
nameCurrrencyCountries = pd.DataFrame(countries[["Name", "Currency"]])
nameCurrrencyCountries
```

## Show only the rows/countries that have more than 2000 GDP (it is in Milliarden USD Bruttoinlandsprodukt)

```python
countries.where(lambda data: data["GDP"]>2000).dropna()
```

## Select all countries where with inhabitants between 50 and 150 Mio

```python
countries.where(lambda data: 50000000 < data["Population"]).where(lambda data: data["Population"] < 150000000).dropna()
```

## Calculate the GDP average (ignore the missing value)

```python
np.average(countries.dropna()["GDP"])
```

## Calculate the GDP average (missing value treated as 0)

```python
np.average(countries["GDP"].replace(math.nan, 0))
```

## Calculate the population density (population/area)  of all countries and add as new column

```python
countries["Density"] = list(map(lambda data: data[1] / data[2], countries.values))
countries
```

## Sort by country name alphabetically

```python
countries.sort_values("Name")
```

## Create a new data frame from the original where the area is changed: all countries with > 1000000 get BIG and <= 1000000 get SMALL in the cell replaced!

```python
bigSmallCountries = pd.DataFrame(countries)
bigSmallCountries["Area"] = bigSmallCountries["Area"].map(lambda data: "BIG" if data > 1000000 else "SMALL")
bigSmallCountries
```
